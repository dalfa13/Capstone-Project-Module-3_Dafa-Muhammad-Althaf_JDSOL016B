#### **Take Out Error Value**
---


```python
import pandas as pd

# 1. Drop value 0 di kolom 'Price'
UC_cleaned = UC[UC['Price'] != 0]

# 2. Drop outlier pada kolom 'Price' menggunakan metode IQR
Q1_Price = UC_cleaned['Price'].quantile(0.25)
Q3_Price = UC_cleaned['Price'].quantile(0.75)
IQR_Price = Q3_Price - Q1_Price

# Menentukan batas bawah dan batas atas untuk Price
lower_bound_Price = Q1_Price - 1.5 * IQR_Price
upper_bound_Price = Q3_Price + 1.5 * IQR_Price

# Menampilkan hasil IQR
print(f'IQR Price: {IQR_Price}')
print(f'Limit Bawah Price: {lower_bound_Price}')
print(f'Limit Atas Price: {upper_bound_Price}')

# Filter outlier pada Price
UC_cleaned = UC_cleaned[(UC_cleaned['Price'] >= lower_bound_Price) & (UC_cleaned['Price'] <= upper_bound_Price)]

# 3. Drop value 'Other' di kolom 'Type'
UC_cleaned = UC_cleaned[UC_cleaned['Type'] != 'Other']

# 4. Drop value 'Other' di kolom 'Make'
UC_cleaned = UC_cleaned[UC_cleaned['Make'] != 'Other']

# 5. Drop outlier pada kolom 'Mileage' menggunakan metode IQR
Q1_Mileage = UC_cleaned['Mileage'].quantile(0.25)
Q3_Mileage = UC_cleaned['Mileage'].quantile(0.75)
IQR_Mileage = Q3_Mileage - Q1_Mileage

# Menentukan batas bawah dan batas atas untuk Mileage (menggunakan 3 * IQR)
lower_bound_Mileage = Q1_Mileage - 1.5  * IQR_Mileage
upper_bound_Mileage = Q3_Mileage + 1.5 * IQR_Mileage

UC_cleaned = UC_cleaned[(UC_cleaned['Mileage'] >= lower_bound_Mileage) & (UC_cleaned['Mileage'] <= upper_bound_Mileage)]

# 6. Drop baris yang duplikat
UC_cleaned = UC_cleaned.drop_duplicates()

# 7. Menghitung jumlah dan persentase data yang dihapus
rows_dropped = len(UC) - len(UC_cleaned)
percentage_dropped = (rows_dropped / len(UC)) * 100

print(f"Jumlah baris yang di-drop: {rows_dropped}")
print(f"Persentase data yang di-drop: {percentage_dropped:.2f}%")
```

**Matrix Evaluation**

Ketika menentukan harga beli mobil yang sesuai menggunakan model regresi, pemilihan metrik evaluasi sangat penting agar hasil prediksi akurat dan stabil. **Tiga metrik utama** yang sering digunakan dalam regresi adalah **RMSE (Root Mean Squared Error) dan MAE (Mean Absolute Error). Berikut adalah alasan mengapa masing-masing metrik ini relevan untuk menilai model yang memprediksi harga mobil bekas:  

---

**1. RMSE (Root Mean Squared Error) â€“ Mengutamakan Error Besar**  

**Alasan menggunakan RMSE:**  
âœ… **Sensitif terhadap outlier**: RMSE mengkuadratkan error sebelum mengambil rata-rata, sehingga memberikan **penalti lebih besar untuk kesalahan besar**. Ini berguna karena dalam harga mobil, terdapat variasi harga tinggi (contoh: mobil mewah dan mobil ekonomi).  
âœ… **Baik untuk model dengan harga tinggi**: Jika kita ingin menghindari kesalahan besar dalam prediksi harga mobil mahal, RMSE adalah metrik yang tepat karena error besar akan sangat berpengaruh pada hasil akhir.  
âœ… **Menjaga stabilitas harga jual**: Dengan menggunakan RMSE, kita dapat meminimalkan kesalahan besar dalam harga jual, sehingga mengurangi risiko menetapkan harga terlalu rendah atau terlalu tinggi.  

ðŸ“Œ **Kelemahan**: Karena lebih sensitif terhadap outlier, RMSE bisa lebih tinggi jika terdapat harga ekstrem dalam dataset.  

---

 **2. MAE (Mean Absolute Error) â€“ Memastikan Error Rata-rata**  

**Alasan menggunakan MAE:**  
âœ… **Lebih mudah diinterpretasikan**: MAE menunjukkan rata-rata kesalahan dalam unit yang sama dengan target (misalnya, jika MAE = 5,000, berarti prediksi harga mobil rata-rata meleset sebesar $5,000).  
âœ… **Tidak terlalu dipengaruhi oleh outlier**: Berbeda dengan RMSE, MAE tidak mengkuadratkan error, sehingga kesalahan besar dan kecil memiliki dampak yang sama.  
âœ… **Cocok untuk rentang harga yang luas**: Dalam pasar mobil bekas, harga bisa bervariasi dari $5,000 hingga $100,000. MAE memberikan gambaran yang lebih realistis tentang rata-rata kesalahan prediksi.  

ðŸ“Œ **Kelemahan**: Tidak memberi penalti lebih besar pada error besar, sehingga bisa kurang efektif dalam mendeteksi outlier dibandingkan RMSE.  

---

 **Kesimpulan**  
Untuk menentukan harga beli mobil yang sesuai, kita menggunakan **kombinasi dari ketiga metrik ini**:  
ðŸ“Œ **RMSE** â†’ Menghindari kesalahan besar dalam harga jual.  
ðŸ“Œ **MAE** â†’ Memberikan gambaran rata-rata kesalahan yang mudah dipahami.  

ðŸ”¹ **Jika ingin memprioritaskan kestabilan harga jual** â†’ Gunakan **RMSE**.  
ðŸ”¹ **Jika ingin meminimalkan kesalahan absolut** â†’ Gunakan **MAE**.  

Idealnya, ketiga metrik ini digunakan bersama untuk mendapatkan pemahaman yang lebih komprehensif tentang performa model dalam menentukan harga mobil yang sesuai. ðŸš—ðŸ’°

```python

# Define Models Algoritm
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=1)
rf = RandomForestRegressor(random_state=1)
ada = AdaBoostRegressor(random_state=1)
gb = GradientBoostingRegressor(random_state=1)
xgb = XGBRegressor(random_state=1)

# Scale models to logaricmic and transform to original (normalize)
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_ada = TransformedTargetRegressor(ada, func=np.log, inverse_func=np.exp)
log_gb = TransformedTargetRegressor(gb, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)

# Scaling data
scaler = StandardScaler()

# Models list
models = [log_lr, log_knn, log_dt, log_rf, log_ada, log_gb, log_xgb]


score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

# Mencari algoritma terbaik berdasarkan tiap metrics
for i in models:
    
    crossval = KFold(n_splits=5, shuffle=True, random_state=1)

    estimator = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', i)
    ])

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
        )

    print(model_cv_rmse, i)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error', 
        error_score='raise'
        )

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())
```

```python
model_names = ['Linear Regression', 'KNN Regressor', 'DecisionTree Regressor', 'RandomForest Regressor', 'AdaBoost Regressor', 'Gradient Boosting Regressor', 'XGBoost Regressor']

# Evaluasi hasil dari 5 kandidat algoritma yang digunakan
pd.DataFrame({
    'Model': model_names,
    'Mean_RMSE': nilai_mean_rmse,
    'Std_RMSE': nilai_std_rmse,
    'Mean_MAE': nilai_mean_mae,
    'Std_MAE': nilai_std_mae,

})
```

Hyperparameter tuning yang akan di lakukan menggunakan model XG Boost Regresion adalah proses untuk mencari nilai optimal dari hyperparameter suatu model machine learning untuk memperbaiki performa model machine learning Ini dilakukan dengan mencoba berbagai nilai hyperparameter dan membandingkan hasil mereka dengan metrik performa seperti akurasi atau F1 score. Hyperparameter tunning dapat di lakukan dengan cara [Tuning XGBoost](https://xgboost.readthedocs.io/en/latest/parameter.html#general-parameters).

```python
# Kedalaman pohon
max_depth = list(np.arange(1, 11))

# Learning rate
learning_rate = list(np.arange(1, 100)/100)

# Jumlah pohon
n_estimators = list(np.arange(100, 201))

# Jumlah baris tiap pohon (% dari total baris train set)
subsample = list(np.arange(2, 10)/10)

# Gamma (min_impurity_decrease)
gamma = list(np.arange(1, 11)) # Semakin besar nilainya, semakin konservatif/simpel modelnya

# Jumlah feature yang digunakan untuk tiap pohon (% dari total kolom train set)
colsample_bytree = list(np.arange(1, 10)/10)

# Alpha (regularization)
reg_alpha = list(np.logspace(-3, 1, 10)) # Semakin besar nilainya, semakin konservatif/simpel modelnya


# Hyperparam space XGboost
hyperparam_space_xgb = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample,
    'model__gamma': gamma,
    'model__colsample_bytree': colsample_bytree,
    'model__reg_alpha': reg_alpha
}
```

```python
# Benchmark model dengan hyperparameter tuning
xgb = XGBRegressor(random_state=1, verbosity=0)

# Membuat algorithm chains
estimator_xgb = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', xgb)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=1)

# Hyperparameter tuning
random_xgb = RandomizedSearchCV(
    estimator_xgb, 
    param_distributions = hyperparam_space_xgb,
    n_iter = 50,
    cv = crossval, 
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error'], 
    n_jobs = -1,
    refit = 'neg_root_mean_squared_error', # Hanya bisa memilih salah stau metric untuk optimisasi
    random_state = 1  
)
```

```python
# Fitting data training untuk mencari parameter terbaik
random_xgb.fit(X_train, y_train)
```

```python
# Melihat hasil tuning dalam bentuk dataframe. Diurutkan berdasarkan RMSE dan MAE
pd.DataFrame(random_xgb.cv_results_).sort_values(by=['rank_test_neg_root_mean_squared_error', 'rank_test_neg_mean_absolute_error']).head()
```

```python
# Cek skor dan params terbaik
print('XGBoost')
print('Best_score:', random_xgb.best_score_)
print('Best_params:', random_xgb.best_params_)
```

### Predict To Test Set Whit The Tuned Model

```python
# Model XGBoost
model = {'XGB': XGBRegressor(random_state=1)}

# Define model terhadap estimator terbaik
xgb_tuning = random_xgb.best_estimator_

# Fitting model
xgb_tuning.fit(X_train, y_train)

# Predict test set
y_pred_xgb_tuning = xgb_tuning.predict(X_test)

# Simpan nilai metrics RMS dan MAE setelah tuning
rmse_xgb_tuning = np.sqrt(mean_squared_error(y_test, y_pred_xgb_tuning))
mae_xgb_tuning = mean_absolute_error(y_test, y_pred_xgb_tuning)

score_after_tuning = pd.DataFrame({'RMSE': rmse_xgb_tuning, 'MAE': mae_xgb_tuning}, index=model.keys())
score_after_tuning
```
